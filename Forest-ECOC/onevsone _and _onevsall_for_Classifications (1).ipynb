## Build Perceptron model :

```python
class Perceptron:
    def __init__(self, name ="Perceptron"):
        self.w : np.ndarray
        self.lossHist : list = []
        self.name = name
        self.ls : float
        self.Tmax = 1000

    def fit(self, X, y, verbose=True):
        m , d = X.shape
        self.w = np.random.rand(d)
        self.lossHist.append(self.loss(X, y))
        self.ls = self.loss(X, y)

        t = 0

        while(self.ls != 0):
            if t >= self.Tmax : break
            for i, xi in enumerate(X):
                if self.hs(xi)*y[i] < 0:
                    self.w += y[i] * xi
                    self.ls = self.loss(X, y)

                t += 1
                if t%10 ==0:
                    self.lossHist.append(self.ls)
                    if verbose:
                        print(f"iter : {t} ,loss : {self.ls}")
                    if self.ls == 0 : break

    def val_data(self, X, y, first=True):

        yp = y.copy()

        c1 = np.unique(yp)[0]

        if first:
            yp = np.where(yp==c1, -1, 1)
        else:
            yp = np.where(yp==c1, 1, -1)

        return X, yp
    
    def split_classes(self,X, y, target_class):
        
      y_new = np.where(y == target_class, 1, -1)
    
      return X, y_new

    def plot_fitted(self, X, y, tmax=5):
        t = np.linspace(0, tmax, 100)
        y_line = - (self.w[0] + t * self.w[1])/self.w[2]
        #plt.scatter(X[:, 1], X[:, 2], marker="o", c=y, color)
        plt.plot(t, y_line)

    def predict(self, X):
        return self.hs(X)

    def loss(self, X, y):
        return (1/len(X)) * sum(0 if self.hs(xi) == y[i] else 1 for i, xi in enumerate(X))

    def hs(self, xi):
        return np.sign(self.w @ xi)

    def summary(self):
        print(f"======== {self.name} model summary : ", "="*50)
        print(f" weights : {self.w}")
        print(f" loss : {self.ls}")
        print(f"\nplot the loss : ")
        plt.plot(range(0, len(self.lossHist)), self.lossHist)
```

##  ADALINE :

```python
class Adaline(Perceptron):

    def __init__(self, eps=0.2):
        super(Adaline, self).__init__("Adaline")
        self.eps = eps

    def loss(self, X, y):
        # Mean squared error loss function
        return (1 / len(X)) * sum((y[i] - self.hs(xi))**2 for i, xi in enumerate(X))

    def grad(self, X, y):
        # Gradient of the mean squared error loss function
        return (1 / len(X)) * sum(2 * xi * (y[i] - self.hs(xi)) for i, xi in enumerate(X))

    def fit(self, X, y, verbose=True):
        m, d = X.shape
        # Initialize weights with random values
        self.w = np.random.rand(d)

        # Record initial loss
        self.lossHist.append(self.loss(X, y))

        # Initialize the current loss value
        self.ls = self.loss(X, y)

        t = 0

        # Continue updating weights until the gradient norm is below the threshold
        while np.linalg.norm(self.grad(X, y)) > self.eps:

            for i, xi in enumerate(X):
                ei = y[i] - self.hs(xi)
                if ei != 0:
                    # Update weights using the Adaline update rule
                    self.w += ei * xi
                    self.ls = self.loss(X, y)

                t += 1
                # Record loss every 10 iterations
                if t % 10 == 0:
                    self.lossHist.append(self.ls)
                    if verbose:
                        print(f"iter : {t} ,loss : {self.ls}")
                    # Break if loss becomes zero
                    if self.ls == 0:
                        break
```

## Pocket

```python
class Pocket(Perceptron):

    def __init__(self, Tmax=100):
        # Constructor for the Pocket class, inheriting from Perceptron
        super(Pocket, self).__init__("Pocket")
        # Initialize the Pocket class with a maximum number of iterations (Tmax)
        self.Tmax = Tmax

    def fit(self, X, y, verbose=True):
        # Fit method for training the perceptron
        m, d = X.shape  # m: number of samples, d: number of features
        self.w = np.random.rand(d)  # Initialize weights randomly

        # Record the initial loss
        self.lossHist.append(self.loss(X, y))
        self.ls = self.loss(X, y)  # Current loss

        t = 0  # Initialize iteration counter

        # Main training loop
        for j in range(self.Tmax):  # Iterate up to Tmax times
            for i, xi in enumerate(X):  # Iterate through each sample
                # Update weights if the prediction is incorrect
                if self.hs(xi) * y[i] < 0:
                    self.w += y[i] * xi
                    self.ls = self.loss(X, y)  # Update loss after weight update

            t += 1  # Increment iteration counter

            # Print and record loss every 10 iterations
            if t % 10 == 0:
                self.lossHist.append(self.ls)
                if verbose:
                    print(f"iter : {t} ,loss : {self.ls}")

                # Break if the loss becomes zero
                if self.ls == 0:
                    break
```

## Logistic Regression

```python
import numpy as np

class LogisticRegression(Perceptron):
    def __init__(self, lr=0.1, Tmax=4000, epsilon=0.1):
        self.lr = lr
        self.Tmax = Tmax
        self.epsilon = epsilon
        self.w = None
        self.t = None
        self.Ls = None

    def phi(self, x):
        return 1 / (1 + np.exp(-x))

    def hs(self, x):
        return self.w.T @ x

    def ls(self, x, y):
        return -np.log(self.phi(y * self.hs(x)))

    def loss(self, X, Y):
        n = len(X)
        error = [self.ls(X[i], Y[i]) for i in range(n)]
        return np.sum(error) / n

    def gradient(self, X, Y):
        n = len(X)
        d = len(self.w)
        dL = np.zeros(d)
        for i in range(n):
            dL += self.phi(-Y[i] * self.hs(X[i])) * (-Y[i] * X[i])
        return dL / n

    def fit(self, X, Y,verbose=True):
        self.t = 0
        self.w = np.zeros(X.shape[1])
        self.Ls = self.loss(X, Y)

        while np.linalg.norm(self.Ls) > self.epsilon and self.t < self.Tmax:
            #if self.t % 100 == 0:
                #print("iter:", self.t, "\t| empirical loss: ", "{0:.6f}".format(self.Ls))
            self.w -= self.lr * self.gradient(X, Y)
            self.Ls = self.loss(X, Y)
            self.t += 1

        return self.w
    def plot_fitted1(self, X, y, tmax=5):
        t = np.linspace(0, tmax, 100)
        y_line = - (self.w[0] + t * self.w[1])/self.w[2]
        #plt.scatter(X[:, 1], X[:, 2], marker="o", c=y, color)
        plt.plot(t, y_line)
```
